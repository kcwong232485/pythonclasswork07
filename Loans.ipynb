```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
np.set_printoptions(precision=2)
%matplotlib inline
```

```python
df = pd.read_csv('loan.csv')
# X = list(zip(df.ApplicantIncome, df.LoanAmount))
# df[['Col1','Col2']].to_numpy()
X = df[['ApplicantIncome','LoanAmount']].to_numpy()
plt.scatter(X[:,0],X[:,1], s=30, c='b')
plt.show()
```

```python
# Use Scikit-Learn
from sklearn.cluster import KMeans
Kmean = KMeans(n_clusters=3)
Kmean.fit(X)
```

```python
# Finding the centroid
cc = Kmean.cluster_centers_
cc
```

```python
plt.scatter(X[:,0], X[:,1], c=Kmean.labels_, cmap="viridis")
color = ['g', 'r', 'b']
for i in range(3):
    plt.scatter(cc[i][0],cc[i][1],s=200, c=color[i],marker='s')
plt.xlabel('Application Income')
plt.ylabel('Loan amount')
#plt.scatter(cc[0][0],cc[0][1],s=200, c='g',marker='s')
#plt.scatter(cc[1][0],cc[1][1],s=200, c='r',marker='s')
#plt.scatter(cc[2][0],cc[2][1],s=200, c='y',marker='s')
```
